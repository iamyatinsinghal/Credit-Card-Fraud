Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Loading the dataset to Pandas to a Pandas DataFrame
ccdata = pd.read_csv('/content/creditcard.csv')
```

```python
#distribution of legit transactions & fraudulent transactions
ccdata['Class'].value_counts()
```

0 ----> Normal Transaction
1 ----> Fraudulent Transaction
(This dataset is highly unbalanced)

```python
#seperating the data for analysis
legit = ccdata[ccdata.Class == 0]
fraud = ccdata[ccdata.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
ccdata.groupby('Class').mean()
```

under-sampling


build a sample dataset containing similar distribution of normal transactions and fraudulent transactions


No. of fraudulent transaction --> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating two Dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.shape
```

```python
new_dataset.info()
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

this dataset is good as contain equal legit and fraudalent transaction

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features & Targets

```python
X = new_dataset.drop(columns ='Class', axis =1)
Y = new_dataset['Class']
```

```python
print(X)
```

X --> Features

Y --> target value

```python
print(Y)
```

Split the data into training data & testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,stratify=Y,random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training

Logistic Regression

```python
model= LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction =model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Training Accuracy={0:.2f}%'.format(training_data_accuracy*100))
```

```python
#accuracy on test data
X_test_prediction =model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Testing Accuracy={0:.2f}%'.format(test_data_accuracy*100))
```
